### Transform the data frame
* `.to_numpy()`: convert data from a DataFrame or a Series to a NumPy array
* `.T`: transpose of the data frame, the columns by indexes and indexes by columns
* `.sort_values("area", ascending=False)`: sort by the values of a specific column in descending order

```python
df_sales_prediction_colombia.to_numpy()
```

```python
type(df_sales_prediction_colombia.to_numpy())
```

```python
df_sales_prediction_colombia.T.head()
```

```python
df_sales_prediction_colombia.sort_values("area", ascending=False).head()
```

### Filter data
* `df["column.name"]`: array with column data
* `df["column.name"].value_counts()`: of the indicated series counts the unique values

##### iloc and loc
* `df[0:4, ["colum1.name","column2.name"]]` : the first four records of the columns
* `df[0:4, 0:3]`: the first four records of the first three columns

```python
df_sales_prediction_colombia["tiempodeconstruido"]
```

```python
df_sales_prediction_colombia["tiempodeconstruido"].value_counts()
```

```python
df_sales_prediction_colombia.loc[0:4, ["antiguedad_original","tiempodeconstruido"]]
```

```python
df_sales_prediction_colombia.iloc[0:4,0:3]
```

### Filter by conditions
* `df[df["column.name"] == "condition"]`: filter the values of a column that is guilty of the condition

```python
df_sales_prediction_colombia[df_sales_prediction_colombia["antiguedad_original"] == "Entre 10 y 20 años"].sort_values("area", ascending=False)
```

```python
df_sales_prediction_colombia[df_sales_prediction_colombia["area"] > 80].head()
```

### Do operations to get a new column
* `df_2 = df.copy() `: create a copy of the original data frame

```python
df_2 = df_sales_prediction_colombia.copy()
```

```python
cm_km2 = df_2["area"] * 1000
```

```python
df_2["area cm²"] = cm_km2
df_2.head()
```
